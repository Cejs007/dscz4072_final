```python
labely = pd.read_csv(f"data/Training_set.csv")
labely
```

```python
skupiny = labely.groupby("label").groups
skupiny
```

```python
X = []
y = []

for druh in list(skupiny.keys())[:3]:
  nazvy_souboru = labely.loc[skupiny[druh], "filename"].values
  for soubor in nazvy_souboru:
    X.append(Image.open(f"data/train/{soubor}"))
    y.append(druh)
```
